```python
import heapq

# Goal State Check
def is_goal(self):
  return self.state == self.goal_state

# PuzzleState Class
class PuzzleState:

  def __init__(self, state, parent=None, move=0):
        self.state = state
        self.parent = parent
        self.move = move
        self.g = 0
        self.h = 0

  # Comparison Methods
  def __lt__(self, other):
        return (self.g + self.h) < (other.g + other.h)

  def __eq__(self, other):
        return self.state == other.state

  def __hash__(self):
        return hash(str(self.state))

  def is_goal(self):
        return self.state == self.goal_state

  def h1(state):
    return 0

  def h2(state):
    displaced = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0 and state[i][j] != PuzzleState.goal_state[i][j]:
                displaced += 1
    return displaced

  def h3(state):
    distance = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0:
                value = state[i][j]
                goal_row, goal_col = divmod(value - 1, 3)
                distance += abs(i - goal_row) + abs(j - goal_col)
    return distance

  def h4(state):
    # Custom heuristic function where h(n) > h*(n)
    return 2 * h3(state)

  # Finding the Empty Tile
  def find_empty(self):
        for i in range(3):
            for j in range(3):
                if self.state[i][j] == 0:
                    return i, j

  def generate_children(self):
        children = []
        empty_row, empty_col = self.find_empty()
        moves = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        for dr, dc in moves:
            new_row, new_col = empty_row + dr, empty_col + dc
            if 0 <= new_row < 3 and 0 <= new_col < 3:
                new_state = [row[:] for row in self.state]
                new_state[empty_row][empty_col], new_state[new_row][new_col] = \
                    new_state[new_row][new_col], new_state[empty_row][empty_col]
                children.append(PuzzleState(new_state, parent=self, move=self.move + 1))
        return children

  # Goal State Definition
  goal_state = [[1, 2, 3],
                  [4, 5, 6],
                  [7, 8, 0]]

# Heuristic Functions
def h1(state):
    return 0

def h2(state):
    displaced = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0 and state[i][j] != PuzzleState.goal_state[i][j]:
                displaced += 1
    return displaced

def h3(state):
    distance = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0:
                value = state[i][j]
                goal_row, goal_col = divmod(value - 1, 3)
                distance += abs(i - goal_row) + abs(j - goal_col)
    return distance

def h4(state):
    # Custom heuristic function where h(n) > h*(n)
    return 2 * h3(state)


# A Search Algorithm*
def astar_search(initial_state, heuristic):
    open_list = []
    closed_set = set()
    heapq.heappush(open_list, initial_state)

    while open_list:
        current_state = heapq.heappop(open_list)
        if current_state.is_goal():
            return current_state

        closed_set.add(current_state)

        for child_state in current_state.generate_children():
            if child_state in closed_set:
                continue

            child_state.g = current_state.g + 1
            child_state.h = heuristic(child_state.state)

            if child_state not in open_list:
                heapq.heappush(open_list, child_state)
            elif child_state.g < current_state.g:
                heapq.heappush(open_list, child_state)

    return None


# Example usage:
start_state = [[2, 8, 3],
               [1, 6, 4],
               [7, 0, 5]]

initial_state = PuzzleState(start_state)
goal_state = astar_search(initial_state, h4)

if goal_state:
  print("Solution found in", goal_state.move, "moves.")
  print("Solution Path:")
  while goal_state:
      for row in goal_state.state:
        print(row)
        print()
        goal_state = goal_state.parent
else:
  print("No solution found.")
```
