```python
# -*- coding: utf-8 -*-
"""1.3  AIML assignment 3 (Hill climbing)

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1MMmS5rWkIOIUXuSHGEyys4SjrVA2OouT

##**Assignment 3:-Hill climbing Search algorithm for solving 8 puzzle problem**
**Sushanta Kumar Das.**
**2403res77(IITP001893)**
E M.Tech batch, Jan'24
"""

import numpy as np
import random

# Input <---initial_state & goal_state
goal_state=np.array([['T1','T2','T3'],['T4','T5','T6'],['T7','T8','B']])
print("Goal state is : "+ str(goal_state))

# Function h1(n) = number of tiles displaced from their destined position.
def h1n_count_displacedtiles(start, target):
    # Find indices of non-blank positions in both start and target arrays
    start_indices = np.where(start != "B")
    target_indices = np.where(target != "B")

    # Count the number of displaced tiles excluding "B"
    displaced_count = np.sum(start[start_indices] != target[target_indices])
    return displaced_count

 # Calculate displaced tiles from start to target (for testing)
displaced_count = h1n_count_displacedtiles(initial_state, goal_state)
print("Number of displaced tiles:", displaced_count)

# Function h2(n) = sum of the Manhattan distance of each tile from the goal position.
def h2n_manhattan_distance(start, target):
    distance = 0
    for i in range(3):
        for j in range(3):
            if start[i][j] != 'B':  # Exclude the blank space
                x1, y1 = np.where(start == start[i][j])
                x2, y2 = np.where(target == start[i][j])
                distance += abs(x1 - x2) + abs(y1 - y2)
    return distance

# Calculate manhattan distance from start to target (for testing)
manhattan_distance = h2n_manhattan_distance(initial_state, goal_state)
print("manhattan_distance:", manhattan_distance)

# Function to generate neighbours
def generate_neighbour(state):
    neighbours = []
    b_row, b_col = np.argwhere(state == 'B')[0]  # Find the position of 'B'
    moves = [(1, 0), (-1, 0), (0, 1), (0, -1)]  # Possible moves: down, up, right, left

    for i, j in moves:
        n_row, n_col = b_row + i, b_col + j  # Corrected variable names
        if 0 <= n_row < 3 and 0 <= n_col < 3:
            new_state = np.copy(state)
            # Swap 'B' with the adjacent element
            new_state[b_row, b_col], new_state[n_row, n_col] = \
                new_state[n_row, n_col], new_state[b_row, b_col]
            neighbours.append(new_state)

    return neighbours


# finding best neighbour using h1(n) = number of tiles displaced from their destined position.
def best_neighbour_state(neighbours, goal_state, heuristic):

    best_neighbour = None
    best_neighbour_value = float('inf')

    for neighbour in neighbours:
        #value = h1n_count_displacedtiles(neighbour, goal_state)
        value = heuristic(neighbour, goal_state)
        if value < best_neighbour_value:
            best_neighbour_value = value
            best_neighbour = neighbour
    return best_neighbour, best_neighbour_value


# Hill climbing search algorithm (using heuristic passed as parameter to the function)
def hill_climbing( initial_state, goal_state, heuristic, no_of_tries=100):

    for try_no in range(no_of_tries):

        #configuration of random initial start
        flattened_state = initial_state.flatten()  # Flatten the array
        np.random.shuffle(flattened_state)  # Shuffle the flattened array
        random_initial_state = flattened_state.reshape(initial_state.shape)  # Reshape to original shape
        #random_initial_state=initial_state (for testing )
        #
        state_explored = 0
        current = random_initial_state.copy()

        for iteration in range(100):
            state_explored += 1

            # Calculate the value of current- from function displaced tile position h1(n)
            #current_value = h1n_count_displacedtiles(current, goal_state)
            current_value = heuristic(current, goal_state)

            # Find the best neighbour/or, higest value neighbour/or, lowest heuristics
            neighbours = generate_neighbour(current)
            best_neighbour, best_neighbour_value = best_neighbour_state(neighbours, goal_state, heuristic)

            # Checking: If the best neighbour is better than the current state & whether goal is reached
            if best_neighbour_value <= current_value and best_neighbour_value == 0:
                print("\nSuccess! Goal state reached at try no", try_no+1)
                print("no of states explored.", state_explored)
                print("initial state:")
                print(random_initial_state)
                print("goal state:")
                print(goal_state)
                return

            else:
                current = best_neighbour

        else:
            print("\nFailed! Try -", try_no+1)
            print("initial state:")
            print(random_initial_state)

#-----------------------------------------------------------
## Sample run

initial_state=np.array([['T6','T7','T3'],['T8','T4','T2'],['T1','B','T5']])
print("Initial state is "+str(initial_state)) #initial state randomization is done as many times as tje no. of retries specified in the parameter

print("Running Hill Climbing with first heuristic....... - no of displaced tiles from their destination position")
hill_climbing_output = hill_climbing( initial_state, goal_state, h1n_count_displacedtiles, no_of_tries=100)


print("Running Hill Climbing with second heuristic....... - manhattan distance")
hill_climbing_output = hill_climbing( initial_state, goal_state, h2n_manhattan_distance, no_of_tries=100)







```
